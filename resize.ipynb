```python
import os
import cv2
```

```python
# folder path
dir_path = r'/home/mawerid/hacks/nuclear_hack/termo_nuclear/data'

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

```python
for i, image in enumerate(res):
    img = cv2.imread('data/' + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    height, width = img.shape
    scale = img[width:height, 0:width]
    img = img[0:width, 0:width]
    cv2.imwrite('data_resized_true/image_' + str(i) + '.png', img)
    cv2.imwrite('data_scales/image_scale_' + str(i) + '.png', scale)
```
