```python
import cv2
import pylab
import numpy
from scipy import ndimage

im = cv2.imread("data/15-25_1_0.png")
im = im[0:1024, 0:1024]
pylab.figure(0)
pylab.imshow(im)

gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
_, gray = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
gray = cv2.morphologyEx(gray, cv2.MORPH_OPEN, numpy.ones((3, 3), dtype=int))
gray = cv2.GaussianBlur(gray, (5,5), 0)
maxValue = 255
adaptiveMethod = cv2.ADAPTIVE_THRESH_MEAN_C#cv2.ADAPTIVE_THRESH_MEAN_C #cv2.ADAPTIVE_THRESH_GAUSSIAN_C
thresholdType = cv2.THRESH_BINARY#cv2.THRESH_BINARY #cv2.THRESH_BINARY_INV
blockSize = 5 #odd number like 3,5,7,9,11
C = -3 # constant to be subtracted
im_thresholded = cv2.adaptiveThreshold(gray, maxValue, adaptiveMethod, thresholdType, blockSize, C) 
labelarray, particle_count = ndimage.label(im_thresholded)
print (particle_count)
pylab.figure(1)
pylab.imshow(im_thresholded)
pylab.show()
```
